```python
# import libraries

from linkedin_api import Linkedin
from datetime import datetime
import openpyxl
import pandas as pd

# login information and companies ids

login_lkd = "your-email@example.com" 
password_lkd = "your-password"
company_ids = {
    'example 1': '923237',
    'example 2': '94802270',
    'example 3': '67724919',
    'example 4': '1382564',
    'example 5': '664118',
    'example 6': '74912384'
}

# auth

api = Linkedin(login_lkd, password_lkd)

# search linkedin followers
def get_followers_count(company_id):
    profile = api.get_company(company_id)
    follower_count = profile.get('followingInfo', {}).get('followerCount', 0)
    return follower_count

# saving instructions
def save_to_excel(data, file_name='linkedin_followers_data.xlsx'):
    df = pd.DataFrame(data)
    
    try:
        with pd.ExcelWriter(file_name, mode='a', engine='openpyxl') as writer:
            df.to_excel(writer, sheet_name=datetime.now().strftime("%Y-%m-%d"), index=False)
    except FileNotFoundError:
        with pd.ExcelWriter(file_name, engine='openpyxl') as writer:
            df.to_excel(writer, sheet_name=datetime.now().strftime("%Y-%m-%d"), index=False)

# collecting and saving data
def main():
    data = []
    today = datetime.now().strftime("%Y-%m-%d")
    
    for company_name, company_id in company_ids.items():
        followers = get_followers_count(company_id)
        data.append({
            'Date': today,
            'Company': company_name,
            'Followers': followers
        })
    
    save_to_excel(data)

if __name__ == "__main__":
    main()
```
